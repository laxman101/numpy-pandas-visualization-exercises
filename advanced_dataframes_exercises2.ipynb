```python
import pandas as pd
import numpy as np
```

```python
users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

```python
left_df = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
right_df = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

```python

right_df.merge(left_df, how='right', on=None, left_on=None, right_on=None, left_index=False, right_index=False, indicator=False)
```

```python
left_df.merge(right_df, how='outer', on=None, left_on=None, right_on=None, left_index=False, right_index=False, indicator=False)
```

```python
users.merge(roles, left_on='role_id', right_on='id', how='outer', indicator=True)
```
