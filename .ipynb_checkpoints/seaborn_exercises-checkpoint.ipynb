### 4. Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
def get_db_url(db):
    from env import user, password, host
    return f'mysql+pymysql://{user}:{password}@{host}/{db}'
```

```python
url = get_db_url('chipotle')
query = "SELECT * FROM orders"
orders = pd.read_sql(query, url)
orders.head()
```

```python
orders.info()
```

```python
orders['item_price'] = orders.item_price.str.replace('$', '').astype('float')
```

```python
orders.head(6)
```

```python
chicken_bowl = orders[orders.item_name == 'Chicken Bowl']
chicken_bowl.sort_values('quantity', ascending = False)
```

```python
best_sellers = orders.groupby('item_name').quantity.sum().nlargest(4)
best_sellers
```

```python
best_sellers.index
```

```python
revenue = orders[orders.item_name.isin(best_sellers.index)].groupby('item_name').item_price.sum()
revenue
```

```python
revenue = revenue.sort_values(ascending = False).reset_index()
revenue
```

```python
plt.figure(figsize = (8, 5))
sns.barplot(x = 'item_name', y = 'item_price', data = revenue)
plt.title('Revenue by Most Popular Items')

plt.ylabel("Dollar Amounts") # 
plt.xlabel("Revenue")
plt.suptitle("Top 4 selling Items Revenue")
```
