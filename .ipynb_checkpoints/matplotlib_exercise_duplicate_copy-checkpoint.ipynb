```python
# 1. Use matplotlib to plot the following equation.
# y = x2-x+2
 
# You'll need to write the code that generates the x and y points.

# Add an anotation for the point 0, 0, the origin.
```

```python
import matplotlib.pyplot as plt
import math
```

```python
x1 = list(range(0, 20))
y1 = [(xn ** 2 -xn +2) for xn in x1]

plt.plot(x1,y1)
plt.annotate('Origin', xy = (0,0), xytext = (0.25, 15)) 

```

```python
# 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):You can use functions from the math module to help implement some of the equations above.
```

```python
x2 = list(range(0,100))
y2 = [math.sqrt(xn)for xn in x2]
plt.plot(x2,y2)
```

```python
x3 = range(0,100)
y3 = [(xn ** 3) for xn in x3]
plt.plot(x3,y3)
```

```python
x4 = range(0,100)
y4 = [(2 ** xn) for xn in x4]
plt.plot(x4,y4)
```

```python
x5 = range(0,100)
y5 = [(1/(xn+1)) for xn in x5]
plt.plot(x5,y5)
```

```python
# 3 Combine the figures you created in the last step into one large figure with 4 subplots.
```

```python
plt.figure(figsize = (8, 6))

plt.subplot(2,2,1)
plt.plot(x2,y2)



plt.subplot(2,2,2)
plt.plot(x3,y3)



plt.subplot(2,2,3)
plt.plot(x4,y4)



plt.subplot(2,2,4)
plt.plot(x5,y5)
```

```python
# 4 Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.
```

```python
plt.figure(figsize = (8, 6))

plt.subplot(2,2,1)
plt.plot(x2,y2, color = 'r')
plt.title('$ chart  A$')
plt.legend ('A')



plt.subplot(2,2,2)
plt.plot(x3,y3, color = 'b')
plt.title('$ chart  B$')
plt.legend ('B')



plt.subplot(2,2,3)
plt.plot(x4,y4, color = 'g')
plt.title('$ chart  C$')
plt.legend ('C')



plt.subplot(2,2,4)
plt.plot(x5,y5, color = 'y')
plt.title('$ chart  D$')
plt.legend ('D')

```

```python
# 5 Make a new Jupyter notebook named big_o_notation.ipynb
```

```python
import matplotlib.pyplot as plt
from math import log, factorial
n = list(range(1,21))

y1 = [1 for n in n]
y2 = [log(n) for n in n]
y3 = [n for n in n]
y4 = [n*log(n) for n in n]
y5 = [n** 2 for n in n]
y6 = [2**n for n in n]
y7 = [n**n for n in n]
y8 = [n**n for n in n]
```

```python
plt.ylim(0,20)

plt.plot(n, y1, color = 'pink', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y2, color = 'purple', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y3, color = 'orange', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y4, color = 'yellow', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y5, color = 'green', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y6, color = 'red', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y7, color = 'blue', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y8, color = 'black', ls = '--', marker = 'x', ms = '10')

plt.title ("Big O Notation")
plt.xlabel("Elements")
plt.ylabel ("Operations")
plt.legend (['O(1)', 'log(n)', 'O(n)', 'O(n log n)', 'O(n^2)', 'O(2^n)', 'O(n!)']);

```
