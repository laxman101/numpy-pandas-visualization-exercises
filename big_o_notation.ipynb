## Curves to graph:

### y = 0n +1 and label the curve "O(1)"
### y = log(n) and label the curve "O(log n)"
### y = n and label the curve "O(n)"
### y = n * log(n) and label it "O(n log n)"
### y = n2 and label it "O(n^2)"
### y = 2n and label it "O(2^n)"
### y = n! and label it "O(n!)"
### y = nn and label it "O(n^n)"

```python
import matplotlib.pyplot as plt
from math import log, factorial
n = list(range(1,20))

y1 = [1 for n in n]
y2 = [log(n) for n in n]
y3 = [n for n in n]
y4 = [n*log(n) for n in n]
y5 = [n** 2 for n in n]
y6 = [2**n for n in n]
y7 = [n**n for n in n]
y8 = [n**n for n in n]

```

```python
plt.xlim(0,20)
plt.ylim(0,20)

plt.plot(n, y1, color = 'pink', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y2, color = 'purple', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y3, color = 'orange', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y4, color = 'yellow', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y5, color = 'green', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y6, color = 'red', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y7, color = 'blue', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y8, color = 'black', ls = '--', marker = 'x', ms = '10')

plt.title ("Big O Notation")
plt.xlabel("Elements")
plt.ylabel ("Operations")
plt.legend (['O(1)', 'O(log n)', 'O(n)', 'O(n log n)', 'O(n^2)', 'O(2^n)', 'O(n!)', 'O(n^n)']);


```
