### 1. Use matplotlib to plot the following equation.
### y = x^2-x+2
 
### You'll need to write the code that generates the x and y points.

### Add an anotation for the point 0, 0, the origin.

```python
import matplotlib.pyplot as plt
import math
```

```python
x1 = list(range(0, 20))
y1 = [(xn ** 2 -xn +2) for xn in x1]

plt.plot(x1,y1)
plt.annotate('Origin', xy = (0,0), xytext = (0.25, 15)) 

plt.title('$ y = x^2 - x +2$')

plt.xlabel('$x$')
plt.ylabel('$y$')



```

### 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):You can use functions from the math module to help implement some of the equations above.

```python
x2 = list(range(0,100))
y2 = [math.sqrt(xn)for xn in x2]
plt.plot(x2,y2)

plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$y$')
```

```python
x3 = range(0,100)
y3 = [(xn ** 3) for xn in x3]
plt.plot(x3,y3)

plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
```

```python
x4 = range(0,100)
y4 = [(2 ** xn) for xn in x4]
plt.plot(x4,y4)

plt.title('$ y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$y$')
```

```python
x5 = range(0,100)
y5 = [(1/(xn+1)) for xn in x5]
plt.plot(x5,y5)

plt.title('$ y = 1/(x+1)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
```

### 3. Combine the figures you created in the last step into one large figure with 4 subplots.


```python
plt.figure(figsize = (16, 12))

# subplot A
plt.subplot(2,2,1)
plt.plot(x2,y2)

plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$y$')


# subplot B
plt.subplot(2,2,2)
plt.plot(x3,y3)

plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')


# subplot C
plt.subplot(2,2,3)
plt.plot(x4,y4)

plt.title('$ y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$y$')


# subplot D
plt.subplot(2,2,4)
plt.plot(x5,y5)

plt.title('$ y = 1/(x+1)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
```

### 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.figure(figsize = (10, 8))


plt.plot(x2,y2, color = 'r',label='$\sqrt{x}$', alpha=1.0)

plt.plot(x3,y3, color = 'b',label='$x^{3}$', alpha=1.0)

plt.plot(x4,y4, color = 'g',label='$2^{x}$', alpha=1.0)

plt.plot(x5,y5, color = 'y',label='$1/(x+1)$', alpha=1.0)


plt.title('My Equations')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
plt.ylim(0, 12)
plt.xlim(0, 12)



```

### 5. Make a new Jupyter notebook named big_o_notation.ipynb


```python
import matplotlib.pyplot as plt
from math import log, factorial
n = list(range(1,20))

y1 = [1 for n in n]
y2 = [log(n) for n in n]
y3 = [n for n in n]
y4 = [n*log(n) for n in n]
y5 = [n** 2 for n in n]
y6 = [2**n for n in n]
y7 = [n**n for n in n]
y8 = [n**n for n in n]
```

```python
plt.xlim(0,20)
plt.ylim(0,20)

plt.plot(n, y1, color = 'pink', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y2, color = 'purple', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y3, color = 'orange', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y4, color = 'yellow', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y5, color = 'green', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y6, color = 'red', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y7, color = 'blue', ls = '--', marker = 'x', ms = '10')
plt.plot(n, y8, color = 'black', ls = '--', marker = 'x', ms = '10')

plt.title ("Big O Notation")
plt.xlabel("Elements")
plt.ylabel ("Operations")
plt.legend (['O(1)', 'O(log n)', 'O(n)', 'O(n log n)', 'O(n^2)', 'O(2^n)', 'O(n!)', 'O(n^n)']);

```
